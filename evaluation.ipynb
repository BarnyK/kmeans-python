```python
import numpy as np
from kmeans import kmeans
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
colors = [
    "#a9a9a9",
    "#2f4f4f",
    "#8b4513",
    "#2e8b57",
    "#808000",
    "#00008b",
    "#ff8c00",
    "#ffd700",
    "#7fff00",
    "#00ff7f",
    "#4169e1",
    "#dc143c",
    "#00ffff",
    "#00bfff",
    "#0000ff",
    "#ff00ff",
    "#db7093",
    "#f0e68c",
    "#ff1493",
    "#ee82ee",
]
```

```python
plt.rcParams['figure.figsize'] = [8, 8]
def visualize(data, result, centroids, title):
    x = data[:,0]
    y = data[:,1]
    if np.max(result) > 20:
        raise ValueError("I don't have more than 10 colors")
    c = [colors[i] for i in result]
    plt.scatter(x,y,c=c,alpha = 0.6, s=10)
    if centroids is not None:
        plt.scatter(centroids[:,0], centroids[:,1],c="#000000", s=10)
    plt.grid(True)
    plt.title(title)
```

```python
filename = "datasets/artificial/2d-10c.arff"
data, final_centroids, result, metrics, iterations, labels = kmeans(
    filename, True, 9, "random", 1000, "", True
)
```

```python
visualize(data, result, final_centroids, "Found")
```

```python
visualize(data, labels, None, "Reference")
```

```python
# Kmeans++ initialization
filename = "datasets/artificial/2d-10c.arff"
data, final_centroids, result, metrics, iterations, labels = kmeans(
    filename, True, 9, "plusplus", 1000, "", True
)
```

```python
visualize(data, result, final_centroids, "Found")
```

```python
from sklearn.cluster import KMeans
from metrics import calculate_metrics
kmeans = KMeans(n_clusters=9, random_state=0, n_init=1)
clusters = kmeans.fit_predict(data)
metrics = calculate_metrics(data, clusters, labels)
visualize(data, clusters, None, "SKLearn")
metrics
```
